```python
x_str = """
<book>
  <name>전은지</name>
  <school>신성여고</school>
  <year>2023</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
x = et.fromstring(x_str)
```

```python
print(x.tag)
```

```python
x_child = list(x)
print(x_child)
```

```python
name, school, year = x_child
print(name.text)
print(school.text)
print(year.text)
```

```python
name = x.findtext('name')
school = x.findtext('school')
year = x.findtext('year')
print(name)
print(school)
print(year)
```

```python
import requests
a = '야구'    #검색어
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={}'
r = requests.get(url.format(a))
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, "html.parser")
```

```python
prd_link = soup.find('a', attrs = {'class':'match'})
print(prd_link)
```

```python
print(prd_link['href'])
```

```python
url = 'https://search.naver.com/search.naver' + prd_link['href']
r = requests.get(url)
```
